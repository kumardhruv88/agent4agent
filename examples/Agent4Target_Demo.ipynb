# Agent4Target Interactive Demo

This notebook demonstrates how to initialize the Agent4Target workflow and run a forward pass to aggregate multimodal evidence for a therapeutic target.

```python
# Install dependencies if running in Colab or a fresh environment
# !pip install pydantic langgraph typer rich
```

### 1. Import Workflow and Schema

```python
import sys
import os
# Ensure the agent4target package is in the path
sys.path.append(os.path.abspath('..'))

from agent4target.orchestrator.workflow import build_workflow
from agent4target.schema.evidence import TargetRequest
```

### 2. Initialize the Pipeline
Compile the LangGraph state machine deterministically.

```python
app = build_workflow()
print("Workflow compiled successfully!")
```

### 3. Run a Forward Pass
Evaluate the therapeutic potential of a target (e.g., BRAF).

```python
# Initialize state
initial_state = {
    "target": TargetRequest(symbol="BRAF"),
    "raw_evidence": [],
    "unified_evidence": None,
    "scored_target": None,
    "errors": []
}

# Execute graph
result = app.invoke(initial_state)

scored_target = result.get("scored_target")
if scored_target:
    print(f"Target: {scored_target.target.symbol}")
    print(f"Aggregate Score: {scored_target.aggregate_score:.2f}")
    print("\n--- Structured Explanation ---")
    print(scored_target.explanation)
else:
    print("Errors:", result.get("errors"))
```
